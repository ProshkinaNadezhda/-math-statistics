```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
t, N = 150, 100
x = np.array([rd.randint(0, t) for i in range(N)])
y = np.linspace(1, N, N)
x1, x2, x3, x4, x5 = (np.array([0] * N) for _ in range(5))
```

```python
for i in range(N):
    x1[i] = abs(2 * np.mean(x[:i+1]) - t)
    x2[i] = abs((i + 2) * min(x[:i+1]) - t)
    x3[i] = abs((i + 2) / (i + 1) * max(x[:i+1]) - t)
    x4[i] = abs(max(x[:i+1]) + min(x[:i+1]) - t)
    if i:            
        x5[i] = abs(x[0] + sum(x[1:i+1]) / i - t)
```

```python
l = [x1, x2, x3, x4, x5, 'θ = 2*x̄', 'θ = Xmin', 'θ = Xmax', 'θ = Xmax + Xmin', 'θ = X1 + sum(X[1:])/(n-1)']
for i in range(5):
    plt.plot(y, l[i], label = l[i+5])
plt.legend(loc = 'center right', fancybox=True,  bbox_to_anchor=(1, 0.6), fontsize=9)
```
